```python
#fix the format of genres 
def fix_formats(obj):
    # Create an empty list to store the names extracted from the input object
    lst = []

    # Use ast.literal_eval to safely evaluate the input object as a literal expression
    for i in ast.literal_eval(obj):
        # Access the "name" key of the current element and append it to the list
        lst.append(i["name"])

    # Return the list containing extracted names
    return lst

movies["genres"] = movies["genres"].apply(fix_formats)
movies["keywords"] = movies["keywords"].apply(fix_formats)
```

```python
def fix_formats2(obj):
    # Create an empty list to store the names extracted from the input object
    lst = []

    # Counter to keep track of the number of elements processed
    counter = 0

    # Use ast.literal_eval to safely evaluate the input object as a literal expression
    for i in ast.literal_eval(obj):
        # Check if the counter is less than 3
        if counter != 3:
            # Access the "name" key of the current element and append it to the list
            lst.append(i["name"])
            
            # Increment the counter
            counter += 1
        else:
            # If counter reaches 3, break out of the loop
            break

    return lst
movies["cast"] = movies["cast"].apply(fix_formats2)
```

```python
def fetch_director(obj):
    # Create an empty list to store the names of directors extracted from the input object
    lst = []

    # Use ast.literal_eval to safely evaluate the input object as a literal expression
    for i in ast.literal_eval(obj):
        # Check if the "job" key of the current element is "Director"
        if i["job"] == "Director":
            # Access the "name" key of the current element and append it to the list
            lst.append(i["name"])
            
            # Break out of the loop after finding the first director (assuming only one director is needed)
            break

    # Return the list containing extracted director names
    return lst
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
#convert overview in to list
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
#we have to apply one transformation on the feature to convert in to one word
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#now we will concate the gener keywords cast and crew and make new feature tags
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["crew"]
```

```python
#now we need now data frame id title tags
data = movies[["movie_id","title","tags"]]
```

```python
data.to_csv("movies.csv",index=False)
```

```python
#now convert tages in string
data["tags"] = data["tags"].apply(lambda x:" ".join(x))
```

```python
data.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
st = PorterStemmer()

def stemmer(text):
    """
    A function to perform stemming on input text.

    Parameters:
    - text (str): Input text to be stemmed.

    Returns:
    - str: Stemmed text.
    """
    y = []
    for i in text.split():
        y.append(st.stem(i))
    
    return " ".join(y)

data["tags"] = data["tags"].apply(stemmer)
```

## Text To Vectors

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
# Use TfidfVectorizer
tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words="english"
)
# Transform the user input into a TF-IDF vector
vectors = tfidf.fit_transform(data["tags"]).toarray()
```

```python
tfidf.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:10]
```

```python
def recommendation(movie):
    movie_index = data[data["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]

    for i in movies_list:
        print(data.iloc[i[0]].title)

recommendation("Avatar")
```
